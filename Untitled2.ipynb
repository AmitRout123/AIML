```python
import matplotlib.pyplot as plt

```

```python
x=[2,6,3,8,6,9,4]
y=[4,8,10,2,7,6,9]

plt.plot(x,y,'og-.')
plt.title('line plot')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.grid()
plt.show()
   
```

```python
x=[2,6,3,8,6,9,4]
y=[4,8,10,2,7,6,9]

plt.bar(x,y,label='example')
plt.title('Bar Graph')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend()
plt.show()
   
```

```python
x=[2,6,3,8,6,9,4]
y=[4,8,10,2,7,6,9]
plt.bar(x,y,label='example 1')

x1=[4,7,2,9,8]
y1=[9,2,6,4,1]
plt.bar(x1,y1,label='example 2')

plt.title('Bar Graph')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend()
plt.show()
   
```

```python
x=[2,6,3,8,6,9,4]
y=[4,8,10,2,7,6,9]
plt.plot(x,y,label='example 1',color='c')

x1=[4,7,2,9,8]
y1=[9,2,6,4,1]
plt.bar(x1,y1,label='example 2')

plt.title('Bar Graph')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend()
plt.show()
   
```

```python
x = [2,6,3,8,10]
y = [1,3,5,2,9]
plt.bar(x,y,label='example 1',color='gold')

plt.title('Bar Graph')
plt.annotate('Mango',xy=(5.5,6))
plt.xlabel('x - axis')
plt.ylabel('y- axis')
plt.legend()
plt.show()
   
```

```python
ages=[20,40,56,78,34,56,23,12,98,]
bins=[10,20,30,40,50,60,70,80]

plt.hist(ages,bins,histtype='bar',rwidth=0.8,color='indigo')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y- axis')
plt.show()
   
```

```python
x1=[4,5,7,3,5,2]
y1=[1,4,3,7,6,9]

plt.scatter(x1,y1,s=300,marker='*',color='orange')
plt.title('Scatter plot')
plt.xlabel('x - axis')
plt.ylabel('y- axis')
plt.show()
   
```

```python
#pie chart

slices=[10,5,8,3,9]
activities=['Phone','Eating','sleeping','studying','playing']
cols=['b','gold','r','m','g']

plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,
        explode=(0,0,0,1,0,0),autopct='%1.2f%%',radius=1.5)
plt.title('pie chart')
plt.show()

```
